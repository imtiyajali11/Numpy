```python
# find out the index of element, who's value is equal to given value.


ma = [[10,20,30],
      [20,50,40],
      [70,80,20]
      ]
ma

value = 70
```

```python
for i in range(len(ma)):
    #print(i,ma[i])
    for j in range(len(ma[i])):
        #print(j)
        print(i,j, '=', ma[i][j])    # to find out the postion and value of the matrix
    
        
```

```python
# to find the position of value = 20


for i in range(len(ma)):
    for j in range(len(ma[i])):
        if ma[i][j] == value:
            print(i,j)
```

### Question: Find out the 2 values who's some equals to given value



```python
ma = [10,20,30,
      20,50, 40,
      70, 80,20]

value = 100
```

```python
l = []

for i in ma:
    for j in ma:
        if i+j == value:
            if sorted([i,j]) not in l:
                l.append(sorted([i,j]))
print(l)
```
