## Measure of dispursion

### Variance

```python
import numpy as np
```

```python
samplea = np.random.randint(1,100, 100000)
```

```python
sampleb = np.random.randint(1,100, 100000)
```

```python
# find mean, median, var,std, cv, z-score
```

```python
# mean 

samplea
```

```python
meana = np.mean(samplea)
meana
```

```python
meanb = np.mean(sampleb)
meanb
```

```python
mediana = np.median(samplea)
medianb = np.median(sampleb)
mediana,medianb
```

```python
vara = np.var(samplea)
varb = np.var(sampleb)

vara, varb
```

```python
stda =np.std(samplea)
stdb =np.std(sampleb)

stda,stdb
```

```python
cva = (stda/meana) * 100
cvb = (stdb/meanb) * 100

cva,cvb
```
