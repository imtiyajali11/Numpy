# - pro = (sample_mean- popu_mean)/(sample_std/(sqrt(n)))

```python
import pandas as pd
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/danielmoralesp/650bf482113d05e75f283607637ea11b/raw/71b4ebd2472767662ffbd23fa35ec7ad6f5f92a6/heights.csv")
```

```python
sample = df.sample(20, random_state= 42)
```

```python
sample.head()
```

```python
from scipy.stats import stats
```

```python
sample_mean = sample['height'].mean()
sample_mean
```

### - Hn = Population height mean = sample height (Xb == Mu)
 ### - H1 : Population height mean is not equals to sample height mean (Xb != Mu)

```python
tstats, pval = stats.ttest_1samp(df['height'],sample_mean)
```

```python
tstats
```

```python
pval
```

```python
if pval < 0.05:
    print('reject Null Hypothesis')

else:
    print('failed to reject null hypothesis')
```
