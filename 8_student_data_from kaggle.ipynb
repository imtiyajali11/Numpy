```python
tutor = studentsdf[studentsdf['Tutoring']==1]
no_tutor = studentsdf[studentsdf['Tutoring']==0]
```

```python
tutor_sample = tutor.sample(100)
no_tutor_sample = no_tutor.sample(100)

```

```python
from scipy.stats import stats
```

```python
ttest, pval = stats.ttest_ind(tutor_sample['GradeClass'], no_tutor_sample['GradeClass'], equal_var= False)
```

```python
pval
```

# - H_null= totoring and non tutoring students hava same marks
# - H_alt = tutoring and non tutoring students does not have same marks

```python
if pval < 0.05:
    print('Reject null hypothesis')

else:
    print(' Failed to reject null Hypothesis')
```
