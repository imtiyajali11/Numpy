```python
sample = df.sample(37)
```

```python
sample.describe()
```

```python
import numpy as np
```

```python
# prove that the mean of the population and sample are equal


sample_mean = sample['weight'].mean()
n = len(sample)
population_mean = df['weight'].mean()
population_std= df['weight'].std()
```

```python
sample_mean, n, population_mean, population_std
```

```python
ztest = (sample_mean/population_mean)/(population_std/(np.sqrt(n)))
ztest
```

```python
# p =  0.05

# 95% confidence 
```
